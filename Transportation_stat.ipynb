```python
transp = pd.read_csv('tsk19.csv', sep = ";")
```

1)Круговую диаграмму, показывающую суммарный расход бензина по датам по всем направлениям

```python
trp = transp.copy()
trp = trp.groupby('Дата').sum('Расход бензина').plot(kind = 'pie', autopct = '%1.1f%%', figsize = (14,14), y='Расход бензина')
```

2) Гистограмма, показывающая сколько тонн груза прибыло за все время в каждый город

```python
trp = transp.copy()
trp = trp.groupby('Пункт назначения')['Масса груза'].sum()
trp.plot.bar(x = 'Пункт назначения', y = 'Масса груза')
```

3) График, показывающий как менялась масса груза, отправленная из города Липки

```python
trp = transp.copy()
trp = (trp[trp['Пункт отправления'] == 'Липки'])[['Дата','Масса груза']]
trp = trp.groupby('Дата')['Масса груза'].sum()
trp = pd.DataFrame({'Дата':trp.index, 'Масса груза':trp.values})
plt.scatter(trp['Дата'],trp['Масса груза'])
plt.show()
```
